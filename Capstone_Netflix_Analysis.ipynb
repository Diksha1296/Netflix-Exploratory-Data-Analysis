```python
data['seasons']=data['duration'].str.extract(r'(\d+)\s+Seas?on').astype(float)
data['minutes']=data['duration'].str.extract(r'(\d+)\s+min').astype(float)
data['seasons'].fillna(0,inplace=True)
data['minutes'].fillna(0,inplace=True)
```

```python
data.head(20)
```

```python
data['seasons']=data['seasons'].astype(int)
```

```python
data.head(10)
```

```python
data['minutes']=data['minutes'].astype(int)
```

```python
data=data.drop('duration',axis=1)
```

```python
data.head(5)
```

```python
data['date_added'].value_counts()
```

```python
data['date_added']=pd.to_datetime(data['date_added'],errors='coerce')
```

```python
data['post_day']=data['date_added'].dt.day
data['post_month']=data['date_added'].dt.month
data['post_year']=data['date_added'].dt.year
```

```python
data.head(5)
```

```python
data=data.drop(columns=['date_added','release_year'])
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
data['post_day'].unique()
```

```python
day_mode=data['post_day'].mode()[0]
```

```python
data['post_day'].fillna(day_mode,inplace=True)
```

```python
data
```

```python
month_mode=data['post_month'].mode()[0]
```

```python
data['post_month'].fillna(month_mode,inplace=True)
```

```python
year_mmode=data['post_year'].mode()[0]
data['post_year'].fillna(year_mmode,inplace=True)
```

```python
data.isnull().sum()
```

```python
data['post_day'] = data['post_day'].astype(int)
data['post_month'] = data['post_month'].astype(int)
data['post_year'] = data['post_year'].astype(int)
```

```python
data.head(5)
```

```python
data['rating'].value_counts()
```

```python
# Remove rows with incorrect ratings
incorrect_ratings = ['74 min', '84 min', '66 min']
# Use the .isin() method and negate it with '~' for "not in"
data = data[~data['rating'].isin(incorrect_ratings)]

# Verify unique values (optional)
print(data['rating'].unique())
```

# **Data Visualization**

* Create visualizations to represent the distribution of content over different genres.




```python
top_N = 20  # Choose the desired number of top genres
top_genres = data['listed_in'].value_counts().nlargest(top_N).index
plt.figure(figsize=(10, 6))
sns.countplot(y='listed_in', data=data[data['listed_in'].isin(top_genres)], order=top_genres)
plt.title('Distribution of Content Over Top {} Genres'.format(top_N))
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

Visualize the distribution of content across release years.

```python
plt.figure(figsize=(10,8))
sns.histplot(data['post_year'],bins=30,kde=False,color='green')
plt.title('Distribution of Content Across Release Years',color='red')
plt.xlabel('Release year')
plt.ylabel('Count')
plt.show()
```

Explore the geographical distribution of content

```python
data.head(5)
```

```python
top=10
tcountries=data['country'].value_counts().nlargest(top).index
clean_data=data[data['country'].isin(tcountries)]
plt.figure(figsize=(10,10))
sns.countplot(x='country',data=clean_data,order=tcountries,color='red')
plt.title('Geographical Distribution of Content (Top {} countrie)'.format(top),color='Blue')
plt.ylabel('count')
plt.xlabel('countries')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
```

Explore the length of movies or episodes and identify any trends.

```python
movies = data[data['type'] == 'Movie']
tv_shows = data[data['type'] == 'TV Show']
```

**Analyze Movie Lengths:**

```python
# Distribution of movie lengths
plt.figure(figsize=(10, 6))
sns.histplot(movies['minutes'], bins=30, kde=True)
plt.title('Distribution of Movie Lengths')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.show()

# Summary statistics for movie lengths
print(movies['minutes'].describe())
```

**Analyze TV Show Episode Lengths:**

```python
tv_shows['avg_episode_length'] = (tv_shows['minutes'] / tv_shows['seasons']).fillna(0).astype(int)

plt.figure(figsize=(10, 6))
sns.histplot(tv_shows['avg_episode_length'], bins=30, kde=True,color='purple')
plt.title('Distribution of Average TV Show Episode Lengths',color='red')
plt.xlabel('Average Episode Length (Minutes)')
plt.ylabel('Frequency')
plt.show()

print(tv_shows['avg_episode_length'].describe())
```

# **Top Lists and Recommendations:**

Identify and present top-rated movies or TV shows based on user ratings.

```python
movies = data[data['type'] == 'Movie']
tv_shows = data[data['type'] == 'TV Show']
```

```python
top_N = 10
top_movies = movies.sort_values(by=['rating'], ascending=False).head(top_N)
top_shows = tv_shows.sort_values(by=['rating'], ascending=False).head(top_N)
```

```python
plt.figure(figsize=(10,8))
plt.barh(top_movies['title'], top_movies['rating'], color='green')
plt.xlabel('Rating')
plt.ylabel('Movie Title')
plt.title('Top {} Rated Movies'.format(top_N),color='red')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.barh(top_shows['title'], top_shows['rating'], color='coral')
plt.xlabel('Rating')
plt.ylabel('TV Show Title')
plt.title('Top {} Rated TV Shows'.format(top_N),color='navy')
plt.show()
```

# **Genre Trends:**

Analyze trends in the popularity of different genres over time.

# **Correlation Analysis:**

1. Correlation Matrix:
